```python
# Import libraries
from datetime import datetime

import pandas as pd
import numpy as np
```

```python
# Open and Read csv File

file_path = "EURUSD.csv"
headers = ["Date", "Heure", "Open", "High", "Low", "Close", "Volume"]

df = pd.read_csv(file_path, sep = ",", names = headers)
```

```python
# Mean of every columns group by Days

A = df.groupby('Date', as_index=False).agg({'Open':'mean',
                        'High':'mean',
                       'Low' : 'mean',
                       'Close':'mean'})
```

```python
df.join(A,rsuffix='_mean')
```

```python
# Creation of the mean columns
df['Mean'] = df[['Open','High','Low','Close']].mean(axis = 1)
```

```python
# Classification of the day
df['Day'] = pd.to_datetime(df["Date"], format = "%Y.%m.%d").dt.strftime("%a")
```

```python
# The increase
df['Increase'] = (df['Close'] - df["Open"])  
```

```python
df
```

```python
dico = {}
for j in ['Mon','Tue','Wed','Thu','Fri'] :
    a = list(df[df.Day== j].agg({'Mean':['std','max']}).Mean)
    b = df[df.Day == j].agg({'Increase' : 'max'}).Increase
    dico[j] = a + [b]
```

```python
dico
```

```python
line = ["Standard Deviation", "Max Value", "+ Increase"]
```

```python
DF = pd.DataFrame.from_dict(dico)
```

```python
DF.index = line
```
